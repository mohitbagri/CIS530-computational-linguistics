```python
# before second phase
import time
from collections import defaultdict
word_counts = defaultdict(lambda: defaultdict(int))
with open('constrained_results_phase1.txt','r') as f:
    line1 = f.readline() 
    line_split1 = line1.split()
    word1 = line_split1[0]
    Y_pred1 = line_split1[2]
    line_idx = 1
    if not 'O' == Y_pred1[0]: word_counts[word1][Y_pred1[2:]]+=1   
    for line2 in f.readlines():
        line_idx += 1
        line_split2 = line2.split()
        try: 
            word2 = line_split2[0]
            Y_pred2 = line_split2[2]
        except: 
            print("This line has fewer than 3 tokens:",line2,f'({line_idx})')
            continue
            
        if not 'O' == Y_pred2[0]: word_counts[word2][Y_pred2[2:]]+=1   
        if 'I' == Y_pred2[0]:
            if 'O' == Y_pred1[0]:
                print(line_idx,":",line1[:-1],"-----",line2[:-1],f'({line_idx})',word2,word_counts[word2])
            elif not Y_pred2[1:] == Y_pred1[1:]:
                print(line_idx,":",line1[:-1],"-----",line2[:-1],f'({line_idx})',word2,word_counts[word2])
        Y_pred1 = Y_pred2
        line1 = line2
        word1 = word2
     
        
        
# for word in word_counts:
#     S = sum(word_counts[word].values())
#     for label in word_counts[word]: word_counts[word][label] = word_counts[word][label]/S
```

```python
# second phrase
import operator
print("Second phrase")
List_line_splits = []
with open('constrained_results_phase1.txt','r') as f:
    List_line_splits =[ line.split() for line in f.readlines()]

line_split1 = List_line_splits[0]
Y_pred1 = line_split1[2]
word1 = line_split1[0]
line_idx = 1
for line_split2 in List_line_splits[1:]:
    line_idx +=1
    try: 
        Y_pred2 = line_split2[2]
        word2 = line_split2[0]
    except: print("This line has fewer than 3 tokens:",line2,f'({line_idx})')
    if 113<line_idx and line_idx < 117:  print(f"{line_idx}:{word2} {word_counts[word2]}")
    if 'I' == Y_pred2[0]:
        Y_pred2_new = "".join(Y_pred2)
#             if 'O' == Y_pred1[0]:
#                 print(line_idx,":",line1[:-1],"-----",line2[:-1],f'({line_idx})',word2,word_counts[word2])
#             elif not Y_pred2[1:] == Y_pred1[1:]:
#                 print(line_idx,":",line1[:-1],"-----",line2[:-1],f'({line_idx})',word2,word_counts[word2])
        Y_pred2_new_suffix = max(word_counts[word2].items(), key=operator.itemgetter(1))[0]
        if 'O' == Y_pred1[0]:  Y_pred2_new = 'B-'+ Y_pred2_new_suffix
        elif not Y_pred2[1:] == Y_pred1[1:]:
            Y_pred1_new_suffix = max(word_counts[word1].items(), key=operator.itemgetter(1))[0]            
            print('----------')
            S1 =  sum(word_counts[word1].values())
            S2 =  sum(word_counts[word2].values())
            print(word1,word_counts[word1])
            print(word2,word_counts[word2])
            if word_counts[word1][Y_pred1_new_suffix]/S1 > word_counts[word2][Y_pred2_new_suffix]/S2:  
                Y_pred2_new_suffix = Y_pred1[2:]#Y_pred1_new_suffix
            Y_pred2_new = 'I-' + Y_pred2_new_suffix
        if not Y_pred2 == Y_pred2_new:
            print(f'{line_idx}:',line1[:-1],"-----",line2[:-1],word2,word_counts[word2],f"{Y_pred2}(old)-->{Y_pred2_new}(new)")
            word_counts[word2][Y_pred2[2:] ] -=1
            word_counts[word2][Y_pred2_new[2:] ] +=1
            line_split2[2] = Y_pred2_new
     
    Y_pred1 = Y_pred2
    line1 = line2
```

```python
T0 = time.time()
from nltk.corpus import conll2002
test_sents = list(conll2002.iob_sents('esp.testb'))
with open("constrained_results.txt", "w") as out:
    j=0
    for sent in test_sents: 
        for i in range(len(sent)):
            word = sent[i][0]
            gold = sent[i][-1]
            pred = List_line_splits[j][-1]
            j += 1
            out.write("{}\t{}\t{}\n".format(word,gold,pred))
    out.write("\n")


print("Now run: python3.6 conlleval.py constrained_results.txt")
print("Elapsed time:",time.time() - T0)
```
